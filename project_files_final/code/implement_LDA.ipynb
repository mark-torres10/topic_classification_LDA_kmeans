```python
# import packages
import numpy as np
import csv
import gensim

# Gensim
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel

# Plotting tools
import pyLDAvis
import pyLDAvis.gensim  # don't skip this
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
###### import data ######
data_dir = "../data/cleaned_tweets.csv"

# keep tweets in array:
tweet_text = []

with open(data_dir) as csv_file:
    readCSV = csv.reader(csv_file, delimiter = ',')

    # iterate by row
    for row in readCSV:
        # append to array that holds tweets
        tweet_text.append(row)
```

```python
# define a function to separate each tweet into separate words
def sentence_to_words(sentence):
    return(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations
```

```python
# use list comprehension for each tweet to get each tweet as a list of its words
tweet_by_word = [sentence_to_words(tweet) for tweet in tweet_text]
```

```python
###### build LDA model ######
```

```python
# create dictionary
id2word = corpora.Dictionary(tweet_by_word)
```

```python
# create term document frequency
corpus = [id2word.doc2bow(text) for text in tweet_text]
```

```python
# build LDA model
lda_model = gensim.models.ldamodel.LdaModel(corpus = corpus, 
                                           id2word = id2word, 
                                           num_topics = 10, 
                                           random_state = 453, 
                                           update_every = 500, # how often model parameters should be updated 
                                           chunksize = 500, 
                                           passes = 2, 
                                           alpha = 'auto',  # defaults to 1.0 / num_topics as a prior
                                           per_word_topics = True)
```

```python
# print topics and keywords
print(lda_model.print_topics())
doc_lda = lda_model[corpus]
```
